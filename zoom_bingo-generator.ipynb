```python
#Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Template notes

The generator requires at least 25 unique options as strings in a list form to generate a unique 5x5 bingo card.

You can add new things to the template array as you wish, but if the string gets long consider adding a linebreak \n for better cell formatting when generating the table.

```python
template = np.array(['"Um..."', '"Let me just \n share my screen"', 'Someone forgot \n to mute', 'Typo on \n powerpoint', 'Get put into \n the same breakout \n room 3 times', \
                      'Ignores questions \n in the chat', '"Innovate"', 'Speaker has \n a virtual \n background', 'Careers outside \n of academia \n promoted',\
                     "Jessie's cat \n crashes the call", 'Patents mentioned',\
                     '"There is \n some building \n work outside"', 'Anything pandemic \n related mentioned', \
                     '"Can you see \n my slides?"', '"Thanks for \n that lovely \n introduction"',\
                     'Chris vaping', 'Someone requests \n to be made host', \
                    'Mic not catching \n someone speaking', 'SWBio DTP \n praised', '"You are all \n really smart \n people"', \
                     '"The dreaded..."', 'PIPS mentioned', \
                    'Group chat \n meme game \n is strong', 'Failure rate \n of spin-out \n companies mentioned', \
                     '"Biotech"', 'Speaker runs \n out of time', '"For your \n assessment"'])
```

```python
#Create 5x5 bingo card with a free cell in the middle
def CreateCard(template): #Pass template to the function
    #Generate 1d array of all template indices, need to be integer
    randnums = np.arange(0, len(template), 1, dtype = int)
    #Randomly shuffle the indices
    np.random.shuffle(randnums)
    #Use only the first 25 indices
    rands = randnums[0:25]
    #Reshape the random indices to bingo card layout
    rand = rands.reshape((5,5))
    #Create empty bingo card. NB! Use dtype=onject to pass entire string to the array
    bingo_card = np.zeros([5,5], dtype=object)
    
    #Fill bingo card array with the template using the randomised integers
    for i in range(5):
        for j in range(5):
            bingo_card[i,j] = template[rand[i,j]]
            
    #Set the middle cell of the bingo card to free
    bingo_card[2,2] = 'Free'
    
    return bingo_card
```

```python
#Create function to plot the bingo card
def PlotBingoCard(template, save=True): #option to turn off figure saving
    #Create card
    card = CreateCard(template)
    #Create figure, axis
    fig, ax = plt.subplots()
    #Create table
    table = ax.table(cellText=card, cellLoc='center')
    #Scale table wisth, length
    table.scale(2,10) #Looks good with (2,10)
    #Set text font
    table.set_fontsize(50)
    #Turn off x, y-axis on empty plot above table
    ax.axis('off')
    #Save a .png of the bingo card in current directory
    if save == True:
        plt.savefig(r'SWBio_SiS-Bingo.png', dpi=200, bbox_inches='tight')
    #Plot the card, stop figure
    plt.show()
```

### Generate and save your bingo card!

Use the PlotBingoCard function to save a .png of your bingo card to your current directory. The function takes one compulsory argument (the template list/1d array), and one optional argument save=True. By setting save=False you can generate a bingo card without saving a .png to your current directory.

```python
PlotBingoCard(template)
```
